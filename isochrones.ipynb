```python
import requests
import geopandas as gpd
import pydeck as pdk
import pandas as pd
```

```python
#documentation @ http://dev.opentripplanner.org/apidoc/1.5.0/resource_LIsochrone.html

response = requests.get(
    'http://localhost:8062/otp/routers/atlanta/isochrone',
    params= {
        'fromPlace': '33.781961,-84.396635', #lat long pair of starting point
        'mode': 'WALK,TRANSIT',
        'date': '02-24-2022',
        'time': '09:00am',
        'maxWalkDistance': 1500,
        'walkReluctance': 5,
        'minTransferTime': 120,
        'cutoffSec': 1800 #30 minutes
    },
)
```

```python
df = gpd.read_file(response.text)
df = df.explode(index_parts=True)
```

```python
view_state = pdk.ViewState(
    **{"latitude": 33.7, "longitude": -84.3, "zoom": 11, "maxZoom": 16, "pitch": 0, "bearing": 0}
)


polygon_layer = pdk.Layer(
    "PolygonLayer",
    data=df,
    id="geojson",
    opacity=0.2,
    stroked=True,
    get_polygon="geometry.coordinates",
    filled=True,    
    wireframe=True,    
    get_fill_color=[255, 100, 100],
    get_line_color=[0, 0, 0],
    auto_highlight=True,
    pickable=True,
)

r = pdk.Deck(
    polygon_layer,
    initial_view_state=view_state,    
    map_style=pdk.map_styles.LIGHT,
)
r.to_html()
```
